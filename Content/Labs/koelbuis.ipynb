# Theorie

Volgens [Newton's wet van afkoeling](https://nl.wikipedia.org/wiki/Afkoelingswet_van_Newton) is de snelheid waarmee een voorwerp afkoelt evenredig met het verschil in de temperatuur van het voorwerp ($T$) en de omgeving ($T_0$). We kunnen dit schrijven als:

$$
    \dot{Q} = -hA(T(t) - T_0),
$$

waarin 

- $\dot{Q}$ de warmtestroom in $\mathrm{W}$,
- $A$ het oppervlak waardoor koeling optreedt in $\mathrm{m}^2$,
- $h$ de warmteoverdrachtscoëfficiënt in $\mathrm{W/(m^2 K)}$.

Dit levert de differentiaalvergelijking

$$
    C\dot{T} = -hA(T(t) - T_0),
$$

met $C$ de warmtecapaciteit in $\mathrm{J/kg}$. Herschrijven met $\tau = \frac{C}{hA}$ levert:

$$
    -\tau\dot{T} = T(t) - T_0,
$$ (eq:dv1)

met als oplossing:

$$
    T(t) - T_0 = (T(0) - T_0)\text{e}^{-t/\tau}.
$$ (eq:dv1_opl)

We kunnen hieruit dus concluderen dat $\tau$ de karakteristieke tijdsduur is waarin de temperatuur van de buis een factor $\text{e}$ verlaagd ten opzichte van de omgevingstemperatuur. 

Vraag: Controleer bovenstaande door vergelijking [](#eq:dv1_opl) te substitueren in [](#eq:dv1).


Vraag: Controleer de oplossing door te substitueren in de differentiaalvergelijking.
Antwoord: Differentieer T(t). Je krijgt dT/dt = -(T - T0)/tau. Invullen in -tau·dT/dt = T - T0 geeft een identiteit, dus de oplossing klopt.


Straling is niet lineair: Qdot_s = epsilon·sigma·A·(T^4 - T0^4). Voor kleine temperatuurverschillen (DeltaT = T - T0) kun je dit lineariseren, waardoor het in de praktijk bijna lineair lijkt.

Vraag: Laat zien waarom (T0+DeltaT)^4 - T0^4 ≈ 4·T0^3·DeltaT.
Antwoord: Gebruik de binomiale ontwikkeling en laat termen met (DeltaT)^2 en hoger weg bij DeltaT << T0.

Vraag: Hoe groot is de fout in ons temperatuurbereik (ongeveer 60–80 °C)?
Antwoord: De linearisatie onderschat de straling grofweg 20–25% in dit bereik.

## Data analyse

- Bepaal de karakteristieke tijd $\tau$ waarin de temperatuur van buis afneemt. Deze kan verschillend zijn voor de drie bovenstaande configuraties. 
- Bereken hieruit de warmteoverdrachtscoëfficiënt.

````python
# Hier de data en de analyse

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

C_v = 8.8e2
T_omg = 21.7
r = 2.5e-2
height = 10.2e-2
width = 0.27e-2

Volume = np.pi*r**2 * height - (np.pi*(r-width)**2 * height)
massa = 2.7e3 * Volume

ringvlak = np.pi*(r**2 - (r - width)**2)
buitenoppervlak = 2*np.pi*r*height + 2*ringvlak
warmtecapaciteit = massa * C_v

csv_path = "Mkoelbuis.csv"
data = np.loadtxt(csv_path, delimiter=';', encoding='utf-8-sig')
times = data[:, 0]
temps = data[:, 1]

plt.figure()
plt.plot(times, temps, 'k.', label='data')
plt.xlabel('t (s)')
plt.ylabel('T (°C)')
plt.legend()
plt.show()
````

# Resultaten

````python
# Hier de data en de analyse

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def exp_func(t, A, tau, T_omg):
    # A is verschiltemperatuur met omgeving aan start
    # tau is de karakteristieke tijd voor de koeling
    # T_omg is de omgevingstemperatuur
    return (A * np.exp(-t/tau) + T_omg)

p0 = [temps[0] - temps[-1], (times[-1] - times[0]) / 3, temps[-1]]
popt, pcov = curve_fit(exp_func, times, temps, p0=p0, maxfev=10000)

A_exp, tau_exp, T_omg_exp = popt
y_fit = exp_func(times, A_exp, tau_exp, T_omg_exp)

plt.figure()
plt.plot(times, temps, "k.", label="data")
plt.plot(times, y_fit, 'r-',
         label='$T = %0.2f e^{-t/%0.1f} + %0.2f$' % (A_exp, tau_exp, T_omg_exp))
plt.legend()
plt.xlabel('t (s)')
plt.ylabel('T (°C)')
plt.show()

h_exp = warmtecapaciteit / (tau_exp * buitenoppervlak)

print("tau =", tau_exp, "s")
print("T_omg (fit) =", T_omg_exp, "°C")
print("h =", h_exp, "W/m^2 K")
````

# Discussie en conclusie

Uit de exponentiële fit volgt een karakteristieke tijd van ongeveer \(	au \approx 3{,}21\cdot 10^2\,\mathrm{s}\) en een omgevingstemperatuur \(T_0 \approx 27^\circ\mathrm{C}\) (de data is na 630 s nog niet helemaal uitgekoeld, dus \(T_0\) is hier een extrapolatie).

Met \(C \approx 97\,\mathrm{J/K}\) en \(A \approx 1{,}68\cdot 10^{-2}\,\mathrm{m}^2\) geeft dit \(h \approx 18\,\mathrm{W/(m^2\,K)}\). Dat ligt in de orde die je verwacht voor natuurlijke convectie in lucht, waarbij straling typisch een paar \(\mathrm{W/(m^2\,K)}\) kan bijdragen (afhankelijk van emissiviteit).

Verschillen tussen configuraties (met/zonder dop, verticaal/horizontaal) komen vooral door convectie: zonder dop kan lucht in de buis circuleren en draagt het binnenoppervlak extra bij, waardoor de effectieve warmteoverdracht groter wordt (sneller koelen). Straling is aanwezig maar relatief minder gevoelig voor oriëntatie; geleiding speelt vooral binnen het metaal zelf en via contactpunten (grijparm), maar die bijdrage is meestal klein als de contactoppervlakken beperkt en geïsoleerd zijn.
